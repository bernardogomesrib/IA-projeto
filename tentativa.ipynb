```python

import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split;
from sklearn.linear_model import LinearRegression;
from sklearn.metrics import mean_squared_error;
from ydata_profiling import ProfileReport
import plotly.graph_objects as go;
from sklearn.ensemble import RandomForestClassifier


link  = 'archive/Student_Dropout_rate.csv'; #pode usar normalmente
link2 = 'archive/data.csv'; #pode usar normalmente
link3 = 'archive/student_data.csv'; #trocar o nome da coluna Output por Target
link4 = 'archive/dataset1.csv' #no tem separador
dicionario ={"Graduate":0,"Dropout":1,"Enrolled":2}#dicionario para transformar as strings em nmeros
dataset1 = pd.read_csv(link,sep=";");
##dataset2 = pd.read_csv(link2,sep=";");
#dataset3 = pd.read_csv(link3,sep=";");
dataset4 = pd.read_csv(link4);

#dataset3.rename(columns={'Output': 'Target'}, inplace=True)

#Reestruturando o dataset1 para ter as mesmas colunas nas mesmas posies do dataset4
dataset1 = dataset1.reindex(index=dataset4.index, columns=dataset4.columns)



#transformando as strings em nmeros
for i in range(len(dataset1['Target'].values)):
  dataset1['Target'].values[i] = dicionario[dataset1['Target'].values[i]]

"""for i in range(len(dataset2['Target'].values)):
  dataset2['Target'].values[i] = dicionario[dataset2['Target'].values[i]]

for i in range(len(dataset3['Target'].values)):
  dataset3['Target'].values[i] = dicionario[dataset3['Target'].values[i]]
"""
for i in range(len(dataset4['Target'].values)):
  dataset4['Target'].values[i] = dicionario[dataset4['Target'].values[i]]
#fim transformao

#Concatenando todos datasets
#dataset = pd.concat([dataset1, dataset2, dataset3, dataset4], ignore_index=True)
dataset = pd.concat([dataset1, dataset4], ignore_index=True)

#apagando todas as linhas que tem a coluna 'Target' com valor igual a 2
dataset = dataset[dataset['Target'] != 2]
#fim apagar linhas


#transformando o dataframe em um dataframe do pandas
data = pd.DataFrame(data=dataset)

#definindo o que vai ser resultado e o que  informao
y = dataset['Target'] #resultado
x = dataset.drop(['Target',"Course","Application mode","Father's qualification","Mother's qualification","Father's occupation","Mother's occupation","Displaced"],axis=1); #informao



#dividindo os dados em teste e resultado para treino e teste em 80% treino e 20% teste
var_treinamento,var_teste,resp_treinamento,resp_test = train_test_split(x,y,test_size = 0.2,random_state = 5)
#definindo um dataframe para visualizar como foi predito e como foi o resultado
dataframe_results = pd.DataFrame()



#definindo os valores reais no dataframe
dataframe_results['Valor_real'] = resp_test.values
```

```python



# Create an instance of RandomForestClassifier
random_forest = RandomForestClassifier()

# Fit the model
"""
qnt estimators = 117
rmse =  0.19813177016093972
IA acertou 96.07% dos casos.

qnt estimators = 118
rmse =  0.19986221148898362
IA acertou 96.01% dos casos.


qnt estimators = 116
rmse =  0.19986221148898362
IA acertou 96.01% dos casos.

"""

#treinando o modelo
random_forest = RandomForestClassifier(n_estimators=117,random_state=2)
#fim treinamento
random_forest.fit(var_treinamento, resp_treinamento.astype(int))

#fazendo a predio
predicao = random_forest.predict(var_teste)

dataframe_results['valor_predito'] = predicao

fig = go.Figure()
fig.add_trace(go.Scatter(x=dataframe_results.index,
                        y=dataframe_results.Valor_real,
                        mode='lines+markers',
                        name='Valor real',
                        line=dict(color='RED', width=2)))


#Linha com dados preditos
fig.add_trace(go.Scatter(x=dataframe_results.index,
                        y=dataframe_results.valor_predito,
                        mode='lines+markers',
                        name='Valor predito',
                        line=dict(color='BLUE', width=2)))
fig.show()


rmse = (np.sqrt(mean_squared_error(resp_test,predicao)))

print("rmse = ",rmse);


acertos = sum(dataframe_results.valor_predito == dataframe_results.Valor_real)
total_amostras = len(dataframe_results.Valor_real)
precisao_percentual = (acertos / total_amostras) * 100

print(f"IA acertou {precisao_percentual:.2f}% dos casos.")

```
